```python
import numpy as np
import math
import copy
import plotly.offline
import plotly.graph_objs as go
from IPython.display import display,HTML

from simulpy.robobj import robot
from simulpy import view
from simulpy import kin
from simulpy import trajectory

pi = math.pi
```

```python
jointno = 3
jointaxismat = [[0,0,1],[0,1,0],[0,1,0]]
lenmat = [[0,0,1],[0,0,2],[0,0,3]]

robot1 = robot.Robot(jointno,jointaxismat,lenmat)

print(robot1.jointaxismat)
```

```python
view.plotrobot(robot1)
```

```python
thetafinal = [pi/3,pi/3,pi/3]
kin.fwd(robot1,thetafinal)
```

```python
view.plotrobot(robot1)
```

```python
print(robot1.cureffpos)
```

```python
thetainit = [pi/5,pi/3,pi/4]
thetafinal = [pi/3,pi/6,pi/3]
nopoints = 10

trajectorymat = trajectory.calcont(robot1,thetainit,thetafinal,nopoints)
trajectory.plotcont(robot1,trajectorymat,nopoints,1)
```

```python
thetainit = [pi/5,pi/3,pi/4]
thetafinal = [pi/3,pi/6,pi/3]
nopoints = 10

trajectorymat = trajectory.calcdisc(robot1,thetainit,thetafinal,nopoints)
trajectory.plotdisc(robot1,trajectorymat,nopoints,1)
```

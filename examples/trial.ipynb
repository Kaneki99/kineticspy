```python
from kineticspy.robobj import robot
from kineticspy import graphics, kin, trajectory, solidobj

import math
import numpy as np

pi = math.pi
```

```python
jointno = 3 # dont change
jointaxismat = [[0,0,1],[0,1,0],[0,1,0]] #dont change
lenmat = [[0,0,1],[0,0,2],[0,0,3]]

robot1 = robot.Robot(jointno,jointaxismat,lenmat)
```

```python
l = 3
b = 3
h = 2
pos = [3,1,0]

Obstacle1 = solidobj.obstacle(l, b, h, pos)

l = 3
b = 3
h = 2
pos = [-3,-1,0]

Obstacle2 = solidobj.obstacle(l, b, h, pos)
```

```python
viapoints = [[-2.29, 3.98, 2],[-2.29, -3.98, 2],[2.29, -3.98, 2],[2.29, 3.98, 2]]

trajectorymat = trajectory.viapts(robot1, viapoints)

trac_end = trajectory.trajectory_end(robot1, trajectorymat)
trac_joint = trajectory.trajectory_joints(robot1, trajectorymat)
vol = solidobj.volcov(robot1, trajectorymat)
robo1 = graphics.position(robot1, trajectorymat[0])
robo2 = graphics.position(robot1, trajectorymat[len(trajectorymat) - 1])

new_data = Obstacle1 + Obstacle2 + trac_end + trac_joint + vol + robo1 + robo2
```

```python
m = graphics.plot(new_data)
m.show()
```

```python
import math
import numpy as np
from itertools import product
import plotly
import plotly.graph_objs as go

pi = math.pi
```

```python
def jacobcalc(theta1dot, theta2dot, theta1 , theta2):
    xdot = -l1*theta1dot*math.sin(theta1) - l2*(theta1dot + theta2dot)*math.sin(theta1 + theta2)
    ydot = l1*theta1dot*math.cos(theta1) + l2*(theta1dot + theta2dot)*math.cos(theta1 + theta2)
    
    return [xdot, ydot]
```

```python
l1 = 2
l2 = 3

vel_joint1 = np.linspace(-1, 1, 100)
vel_joint2 = np.linspace(-1, 1, 100)

b = vel_joint1, vel_joint2
mat = list(product(*b))

siz = len(mat)

x = np.zeros(siz)
y = np.zeros(siz)
i = 0

while i < siz:
    [x[i], y[i]] = jacobcalc(mat[i][0], mat[i][1], pi/2, pi/4)
    i = i + 1
    
```

```python
end_velocity = go.Scatter(x=x, y=y,)
data = [end_velocity]
plotly.offline.init_notebook_mode(connected=True)

fig = go.Figure(data=data, layout={})
fig.show()
```
